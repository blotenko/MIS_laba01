```python
def is_prime(n):
    """Перевіряє, чи є число простим."""
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def goldbach(n):
    """Знаходить два простих числа, що складаються в суму n (гіпотеза Гольдбаха)."""
    if n <= 2 or n % 2 != 0:
        print("Число повинно бути парне і більше 2.")
        return
    for i in range(2, n // 2 + 1):
        if is_prime(i) and is_prime(n - i):
            print(f"Число {n} можна представити як {i}+{n-i}")
            return

# Приклади використання
goldbach(6)
goldbach(28)
```
